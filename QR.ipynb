```python
import numpy as np
import matplotlib.pyplot as plt
import time

# Define a matrix
A = np.random.rand(50, 50)

# Start timer
start_time = time.time()

# Perform QR Decomposition
Q, R = np.linalg.qr(A)

# End timer
end_time = time.time()

# Calculate execution time
execution_time = end_time - start_time
print(f"QR Decomposition in Python took: {execution_time} seconds")

# Plotting Q and R matrices
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title("Q Matrix")
plt.imshow(Q)
plt.colorbar()

plt.subplot(1, 2, 2)
plt.title("R Matrix")
plt.imshow(R)
plt.colorbar()

plt.show()
```
