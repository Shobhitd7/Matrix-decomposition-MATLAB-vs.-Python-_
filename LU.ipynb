```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import lu
import time

# Define a 50x50 square matrix
A = np.random.rand(50, 50)

# Start timer
start_time = time.time()

# Perform LU Decomposition
P, L, U = lu(A)

# End timer
end_time = time.time()

# Calculate execution time
execution_time = end_time - start_time
print(f"LU Decomposition in Python took: {execution_time} seconds")

# Plotting
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.title("Matrix A")
plt.imshow(A)
plt.colorbar()

plt.subplot(1, 3, 2)
plt.title("L Matrix")
plt.imshow(L)
plt.colorbar()

plt.subplot(1, 3, 3)
plt.title("U Matrix")
plt.imshow(U)
plt.colorbar()

plt.show()
```
